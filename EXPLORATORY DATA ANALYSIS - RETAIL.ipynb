<b style="color:DarkGreen;font-size:20px;">Rel Plot</b><br>
This function provides access to several different axes-level functions that show the relationship between two variables with semantic mappings of subsets. The kind parameter selects the underlying axes-level function to use:

scatterplot() (with kind="scatter"; the default)

lineplot() (with kind="line")

<h2 style="color:green;">API abstraction across visualizations</h2>
There is no universally best way to visualize data. Different questions are best answered by different plots. Seaborn makes it easy to switch between different visual representations by using a consistent dataset-oriented API.

The function <b>relplot()</b> is named that way because it is designed to visualize many different statistical relationships. While scatter plots are often effective, relationships where one variable represents a measure of time are better represented by a line. The <b>relplot()</b> function has a convenient kind parameter that lets you easily switch to this alternate representation:

```python
sns.relplot(
    data=df, kind="line",
    x="Profit", y="Sales", col="Category",
    hue="Category",
    facet_kws=dict(sharex=False),
)
```

<b style="color:DarkGreen;font-size:20px;">Machine Learning Algorithms</b><br>
<b style="font-size:18px;"><li>Decision Tree Regressor</li>
<li>Linear Regression</li>
<li>Random Forest Regressor</li>

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
df2 = df.drop(['Ship Mode','Segment','Country','City','State','Postal Code','Category','Sub-Category','Profit %'],axis=1)
```

```python
x = df2[['Quantity','Cost']]
y = df2['Sales']
```

<b style="color:DarkGreen;font-size:20px;">Decision Tree Regressor Algorithm</b><br>


<b style="color:DarkGreen;font-size:18px;">Model Fitting using Decision Tree Regressor</b><br>


```python
(train_inputs, test_inputs, train_classes, test_classes) = train_test_split(x, y, train_size=0.7, test_size=0.3,random_state=50)
declf = DecisionTreeRegressor()
declf.fit(train_inputs, train_classes)
```

```python
predictions = declf.predict(train_inputs)
```

```python
new_inputs = [[3,512.3580]]
predictions = declf.predict(new_inputs)
predictions
```

<b style="color:DarkGreen;font-size:20px;">Decision Tree Regressor Training and Testing Accuracy</b><br>


```python
print("Accuracy of Training of Decision Tree -",declf.score(train_inputs, train_classes)*100)
print("Accuracy of Testing of Decision Tree -",declf.score(test_inputs, test_classes)*100)
```

<b style="color:DarkGreen;font-size:20px;">Linear Regressor Algorithm</b><br>


```python
from sklearn.linear_model import LinearRegression
linearRegressor= LinearRegression()
linearRegressor.fit(train_inputs, train_classes)
predictions = declf.predict(train_inputs)
```

<b style="color:DarkGreen;font-size:20px;">Model Fitting using Linear Regression</b><br>


```python
lr = [[3,512.3580]]
predictions = linearRegressor.predict(lr)
predictions
```

<b style="color:DarkGreen;font-size:20px;">Linear Regression Training and Testing Accuracy</b><br>


```python
print("Accuracy of Training of  Linear Regression -",linearRegressor.score(train_inputs, train_classes)*100)
print("Accuracy of Testing of Linear Regression -",linearRegressor.score(test_inputs, test_classes)*100)
```

<b style="color:DarkGreen;font-size:20px;">Random Forest Regressor Algorithm</b><br>


<b style="color:DarkGreen;font-size:18px;">Model Fitting using Random Forest Regressor</b><br>


```python
from sklearn.ensemble import RandomForestRegressor  
regressor = RandomForestRegressor(n_estimators = 250, random_state = 82) 
regressor.fit(train_inputs, train_classes)   
```

```python
Y_pred = regressor.predict(train_inputs)   
```

```python
rf = [[3,512.3580]]
predictions = regressor.predict(rf)
predictions
```

```python
print("Accuracy of Training of  Random Forest Regression -",regressor.score(train_inputs, train_classes)*100)
print("Accuracy of Testing of Random Forest Regression -",regressor.score(test_inputs, test_classes)*100)
```
